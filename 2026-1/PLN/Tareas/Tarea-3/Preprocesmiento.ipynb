<h3>Removemos stop words</h3>

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def remove_stopwords(text):
    doc = nlp(text)
    return " ".join(
        token.lemma_.lower()
        for token in doc
        if not token.is_stop
        and token.is_alpha
    )

df_clean = df.copy()
df_clean['reviews.text'] = df_clean['reviews.text'].apply(remove_stopwords)
```

<h1>Extracción de opiniones asociadas a aspectos

```python

df_clean["doc"] = None

mask = df_clean["reviews.text"].notna()
texts = df_clean.loc[mask, "reviews.text"].tolist()
docs = list(nlp.pipe(texts, batch_size=50))
df_clean.loc[mask, "doc"] = pd.Series(docs, index=df_clean.index[mask])
```

```python
from spacy.matcher import Matcher

matcher = Matcher(nlp.vocab)

items = [
    "tv", "tablet", "screen", "speaker", "book", "music", "case", "box",
    "video", "sound", "battery", "price", "game", "cover", "remote",
    "phone", "kindle", "size", "display", "headphone"
]

patterns = [
    [
        {'LOWER': item},
        {'POS': {'IN': ['DET', 'ADJ']}, 'OP': '*'},
        {'POS': 'NOUN', 'OP': '*'},
    ]
    for item in items
]

matcher.add("INTENCION_ITEM", patterns)
```

```python
from collections import defaultdict

def get_matches_by_item(doc):
    if doc is None:
        return {}

    results = defaultdict(list)

    for _, start, end in matcher(doc):
        span = doc[start:end]


        if len(span) <= 1:
            continue

        item = span[0].text.lower()
        results[item].append(span.text)

    return dict(results)


```

```python
df_clean["matches_by_item"] = df_clean["doc"].apply(get_matches_by_item)



```

<h2>Patrones de intensificación

```python
from spacy.matcher import Matcher

matcher = Matcher(nlp.vocab)

pattern = [
    {"POS": "ADV"},
    {"POS": "ADJ"}
]

matcher.add("ADV_ADJ", [pattern])
```

```python
from collections import defaultdict

aspect_opinions = defaultdict(list)
```

```python
WINDOW = 3

for text in df_clean['reviews.text']:
    doc = nlp(text)
    matches = matcher(doc)

    for _, start, end in matches:
        span = doc[start:end]   # ADV + ADJ

        left = max(0, start - WINDOW)
        right = min(len(doc), end + WINDOW)

        context = doc[left:right]

        for token in context:
            if token.text in aspects:
                aspect_opinions[token.text].append(span.text)
```

```python
for aspect, opinions in aspect_opinions.items():
    print(f"{aspect}:")
    for op in opinions:
        print(f'  "{op}"')
```

<h2> Construcciones verbo + adverbio / adjetivo

```python
from collections import defaultdict

aspect_actions = defaultdict(list)
```

```python
for text in df_clean['reviews.text']:
    doc = nlp(text)

    for token in doc:
        # buscamos VERBOS
        if token.pos_ == "VERB":

            # buscamos modificadores del verbo
            modifiers = [
                child for child in token.children
                if child.pos_ in {"ADV", "ADJ"}
            ]

            if not modifiers:
                continue

            # buscamos sujetos u objetos del verbo
            arguments = [
                child for child in token.children
                if child.dep_ in {"nsubj", "dobj", "nsubjpass", "obj"}
            ]

            for arg in arguments:
                if arg.text in aspects:
                    for mod in modifiers:
                        aspect_actions[arg.text].append(
                            f"{token.lemma_} {mod.text}"
                        )
```

```python
for aspect, actions in aspect_actions.items():
    print(f"{aspect}:")
    for act in set(actions):
        print(f'  "{act}"')
```
