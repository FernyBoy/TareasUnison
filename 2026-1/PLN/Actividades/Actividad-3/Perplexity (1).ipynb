# Matrices

## Matriz de adyacencia

```python
import pandas as pd

words = text.split()

bigrams = [(words[i], words[i+1]) for i in range(len(words)-1)]

df_temp = pd.DataFrame(bigrams, columns=['Word-1', 'Word-2'])

matrix = pd.crosstab(df_temp['Word-1'], df_temp['Word-2'])

matrix
```

## Matriz de probabilidad

```python
matrix_percentages = matrix.div(matrix.sum(axis=1), axis=0)

matrix_percentages = matrix_percentages.fillna(0)

matrix_percentages.round(2)
```

## Generador de frases

```python
import pandas as pd
import numpy as np

def generate_frase(df_matrix, start_word, length=5):
    # Asegurarnos de que no usamos la columna 'Suma' para elegir palabras

    frase = [start_word]
    current_word = start_word

    for _ in range(length - 1):
        if current_word in matrix.index:
            # Obtenemos la fila de la palabra actual
            distribution = matrix.loc[current_word]

            # Si todas las frecuencias son 0, nos detenemos o elegimos una al azar
            if distribution.sum() == 0:
                break

            # Elegimos la siguiente palabra basada en las frecuencias (pesos)
            next = np.random.choice(distribution.index, p=distribution/distribution.sum())
            frase.append(next)
            current_word = next
        else:
            break

    return " ".join(frase)

start_word = np.random.choice(matrix.index)
frase_1 = generate_frase(matrix, start_word, length=5)

start_word = np.random.choice(matrix.index)
frase_2 = generate_frase(matrix, start_word, length=5)

print(f"Frase 1: {frase_1}")
print(f"Frase 2: {frase_2}")
```

# Cálculo de perplejidad

## Función de perplejidad:
#### $Pp(W) = \sqrt[N]{\frac{1}{P(w_{1}w_{2}...w_{N})}}$

## Calculamos la cantidad de bigramas totales (nuestra $N$)

```python
bigrams_list = [(words[i], words[i+1]) for i in range(len(words)-1)]

N = len(bigrams_list)
```

## Creamos una función que calcule la perplejidad

```python
def calculate_perplexity(frase):
  # Creamos el arreglo con las palabras de la frase
  words = frase.split()

  # Inicializamos la variable de la probabilidad
  probability = 1;

  # Recorremos el arreglo de palabras multiplicando sus probabilidades
  for index in range(len(words) - 1):
    probability *= (matrix_percentages.at[words[index], words[index + 1]])

  # Retornamos la perplejidad
  return (1/probability)**(1/N)
```

## Calculamos la perplejidad de las 2 oraciones generadas

```python
print(f'Cantidad de bigramas totales (N): {N}')
print(f'\n{50 * "-"}\n')

print(f'Frase 1: {frase_1}')
print(f'Perplejidad: {calculate_perplexity(frase_1)}')

print(f'\n{50 * "-"}\n')

print(f'Frase 2: {frase_2}')
print(f'Perplejidad: {calculate_perplexity(frase_2)}')
```
